```python
df1 = data['host_response_time'].str.get_dummies()
test2 = data['host_is_superhost'].str.get_dummies()
profPicDum =  data['host_has_profile_pic'].str.get_dummies()
hostVerif =  data['host_identity_verified'].str.get_dummies()
neighbourhood = data['neighbourhood_cleansed'].str.get_dummies()
houseType =  data['property_type'].str.get_dummies()
roomType = data['room_type'].str.get_dummies()
bedType = data['bed_type'].str.get_dummies()
guestPic = data['require_guest_profile_picture'].str.get_dummies()
guestPhone = data['require_guest_phone_verification'].str.get_dummies()
cancellation = data['cancellation_policy'].str.get_dummies()
```

```python
df1 = df1.rename(columns={'a few days or more':'responseTime_fewDays', 'within a day':'responseTime_day','within a few hours':'responseTime_fewHours','within an hour':'responseTime_hour'})
df1 = df1.join(test2)
df1.drop('f', axis=1,inplace=True)
df1 = df1.rename(columns={'t':'superhost'})
df1 = df1.join(profPicDum, lsuffix='_df1', rsuffix='_profPicDum')
df1.drop('f',axis=1,inplace=True)
df1 = df1.rename(columns={'t':'host_hasProfilePic'})
df1 = df1.join(hostVerif, lsuffix='_df1', rsuffix='_hostVerif')
df1.drop('f',axis=1,inplace=True)
df1 = df1.rename(columns={'t':'host_id_verified'})
df1 = df1.join(neighbourhood, lsuffix='_df1', rsuffix='_neighbourhood')
df1 = df1.join(houseType, lsuffix='_df1', rsuffix='_houseType')
df1 = df1.join(roomType, lsuffix='_df1', rsuffix='_roomType')
df1 = df1.join(bedType, lsuffix='_df1', rsuffix='_bedType')
df1 = df1.join(guestPic, lsuffix='_df1', rsuffix='_guestPic')
df1.drop('f', axis=1,inplace=True)
df1 = df1.rename(columns={'t':'require_guestPic'})
df1 = df1.join(guestPhone, lsuffix='_df1', rsuffix='_guestPhone')
df1.drop('f', axis=1,inplace=True)
df1 = df1.rename(columns={'t':'require_guestPhone'})
df1 = df1.join(cancellation, lsuffix='_df1', rsuffix='_cancellation')
```
