```python
#Build model from data

def getCols():
    allcols = pd.read_csv(path + "\modelXvars_AMS.csv")
    allcols = allcols.drop(columns = ['Unnamed: 0'])
    allc = allcols.name[allcols.Keep == True].tolist()

    #load also columns in different categories predefined
#    flexcols = pd.read_csv(path + "/flexCols.csv")
#    flexcols = flexcols.drop(columns = ['Unnamed: 0'])["0"].tolist()
#,Keep,name,flex,location,service,type
    flexcols = allcols.name[allcols.flex == True].tolist()
    loccols = allcols.name[allcols.location == True].tolist()
    servicecols = allcols.name[allcols.service == True].tolist()
    typecols = allcols.name[allcols.type == True].tolist()
    return (allc,flexcols, loccols, servicecols, typecols)

def getAreas():
    allcols = pd.read_csv(path + "\modelXvars_AMS.csv")
    return allcols.name[allcols.location == True].tolist()

def getTypes():
    allcols = pd.read_csv(path + "\modelXvars_AMS.csv")
    return allcols.name[allcols.type == True].tolist()
```

```python
#pick one of these categories
#cols = modcols.name[modcols.Keep == True].tolist() #all
import numpy as np

#cols = columns to include in model as a list
#area = areas to consider in model creation, list

def makeModel(cols, area = ""):
#    cols = loccols
    #cols = servicecols
    #cols = typecols
    if (area == ""):
        area = getAreas()

    dMod = pd.read_csv(path + '/cleansed_with_income_nonans_AMS.csv')
    dMod= dMod.drop(columns=['Unnamed: 0'])
    dMod = dMod[dMod['neighbourhood_cleansed'].isin(area)]
    
    dModX = dMod[cols]
   # dModY = dMod.incomeRate
    dModY = dMod.price


    X =  dModX
    Y = dModY

    import statsmodels.api as sm
    from scipy import stats
    #First model to identify 
    X2 = sm.add_constant(X)
    est = sm.OLS(Y, X2)
    est2 = est.fit()
    print(est2.summary())

    #return the whole model
    #est.conf_int() --> dataframe of lower (0) and upper (1) bounds
    #model.pvalues --> p-values as a pandas series
    #model.params --> regression coefficients
    #model.predict(X) --> to apply the model to specific case i.e. to see what price should be for given set up
    return est2


```
